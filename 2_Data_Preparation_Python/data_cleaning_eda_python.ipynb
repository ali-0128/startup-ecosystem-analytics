```python
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
acquisitions        = pd.read_csv(r'acquisitions.csv')
degrees             = pd.read_csv(r'degrees.csv')
funding_rounds      = pd.read_csv(r'funding_rounds.csv')
funds               = pd.read_csv(r'funds.csv')
investments         = pd.read_csv(r'investments.csv')
ipos                = pd.read_csv(r'ipos.csv')
milestones          = pd.read_csv(r'milestones.csv')
objects             = pd.read_csv(r'objects.csv')
offices             = pd.read_csv(r'offices.csv')
people              = pd.read_csv(r'people.csv')
relationships       = pd.read_csv(r'relationships.csv')
```

# **Initial Exploration**

```python
dataset = {
    'acquisitions': acquisitions,
    'degrees': degrees,
    'funding_rounds': funding_rounds,
    'funds': funds,
    'investments': investments,
    'ipos': ipos,
    'milestones': milestones,
    'objects': objects,
    'offices': offices,
    'people': people,
    'relationships': relationships
}

for name, df in dataset.items():
    print("\n" + "=" *150)
    print(f'Dataset: {name}')
    print("=" *150)

    # Display dataset dimensions
    print("\n [1] Shape")
    shape = pd.DataFrame(
        {"N_Rows": [df.shape[0]], "N_columns":[df.shape[1]]}
    )
    display(shape)

    # Check for identical rows
    print("\n [2] Check Duplicated")
    dupl = df.duplicated().sum()
    display(dupl)

    # Check column data types 
    print("\n [3] Data Type")
    columns_type = pd.DataFrame(
        {"Columns": df.columns, "Data Type": df.dtypes}
    )
    display(columns_type)

    # Statistical overview for numeric and categorical data
    print("\n [4] Describe")
    describe_df = df.describe(include= "all").T
    display(describe_df)
    
    # Identify missing values per column
    print("\n [5] Missing Data")
    missing = df.isnull().sum()
    missing_df = pd.DataFrame({
        "Column"       : missing.index,
        "Missing Count": missing.values,
        "Missing %"    : (missing.values / len(df) * 100).round(2)
    })
    missing_df = (
        missing_df[missing_df["Missing Count"] > 0]
        .sort_values("Missing %", ascending=False)
        .reset_index(drop=True)
    )
    display(missing_df if not missing_df.empty else "No missing values")

    # First 10 Rows
    print("\n [6] First Rows")
    display(df.head(10))

    # Last 10 Rows
    print("\n [7] Last Rows")
    display(df.tail(10))
    
    # Summary
    print("\n [8] Summary")
    summary = pd.DataFrame({
        "N_Rows": [df.shape[0]],
        "N_Columns": [len(df.columns)],
        "Memory": [df.memory_usage(deep= True).sum()]
    }) 
    display(summary)
```

# **Data Cleaning**

# **Drop duplicates**

```python
acquisitions.drop_duplicates(inplace=True)      
degrees.drop_duplicates(inplace=True)           
funding_rounds.drop_duplicates(inplace=True)    
funds.drop_duplicates(inplace=True)            
investments.drop_duplicates(inplace=True)       
ipos.drop_duplicates(inplace=True)              
milestones.drop_duplicates(inplace=True)        
objects.drop_duplicates(inplace=True)           
offices.drop_duplicates(inplace=True)          
people.drop_duplicates(inplace=True)            
```

# **Drop columns**

```python
people.drop(columns=['birthplace'], inplace=True)
```

```python
offices.drop(columns=['created_at', 'updated_at'], inplace=True)

offices.drop(columns=['description', 'address1', 'address2', 'zip_code'], inplace=True)

# Filter out invalid coordinates (0,0 = in the ocean)
offices = offices[(offices['latitude'] != 0) | (offices['longitude'] != 0)]
```

```python
objects.drop(columns=['short_description','description','tag_list','parent_id','logo_url',
                      'logo_width','logo_height','homepage_url','domain','twitter_username',
                      'overview','created_by'], inplace=True)
```

```python
milestones.drop(columns=['milestone_code', 'source_url', 'source_description'], inplace=True)
```

```python
ipos.drop(columns=['source_url', 'source_description'], inplace=True)
```

```python
funds.drop(columns=['source_url', 'source_description'], inplace=True)
```

```python
funding_rounds.drop(columns=['pre_money_currency_code', 'post_money_currency_code', 'source_url', 'source_description', 'created_by'], inplace=True)
```

```python
acquisitions.drop(columns=['term_code', 'source_url', 'source_description'], inplace=True)
```

## **Handle missing values**

```python
acquisitions['price_currency_code'].fillna('USD', inplace=True)
acquisitions.dropna(subset=['acquired_object_id'], inplace=True)
```

```python
funding_rounds['raised_currency_code'].fillna('USD', inplace=True)
funding_rounds.dropna(subset=['funded_at'], inplace=True)
```

```python
ipos['raised_currency_code'].fillna('USD', inplace=True)
ipos['valuation_currency_code'].fillna('USD', inplace=True)
ipos.dropna(subset=['object_id'], inplace=True)
```

```python
objects['category_code'].fillna('unknown', inplace=True)
objects['funding_total_usd'].fillna(0, inplace=True)
```

## **Fix data types**

```python
acquisitions['acquired_at'] = pd.to_datetime(acquisitions['acquired_at'], errors='coerce')
acquisitions['created_at']  = pd.to_datetime(acquisitions['created_at'],  errors='coerce')
acquisitions['updated_at']  = pd.to_datetime(acquisitions['updated_at'],  errors='coerce')
```

```python
degrees['graduated_at'] = pd.to_datetime(degrees['graduated_at'], errors='coerce')
degrees['created_at']   = pd.to_datetime(degrees['created_at'],   errors='coerce')
degrees['updated_at']   = pd.to_datetime(degrees['updated_at'],   errors='coerce')
```

```python
funding_rounds['funded_at']    = pd.to_datetime(funding_rounds['funded_at'],    errors='coerce')
funding_rounds['created_at']   = pd.to_datetime(funding_rounds['created_at'],   errors='coerce')
funding_rounds['updated_at']   = pd.to_datetime(funding_rounds['updated_at'],   errors='coerce')

funding_rounds['is_first_round'] = funding_rounds['is_first_round'].astype(bool)
funding_rounds['is_last_round']  = funding_rounds['is_last_round'].astype(bool)
```

```python
funds['funded_at']   = pd.to_datetime(funds['funded_at'],   errors='coerce')
funds['created_at']  = pd.to_datetime(funds['created_at'],  errors='coerce')
funds['updated_at']  = pd.to_datetime(funds['updated_at'],  errors='coerce')
```

```python
investments['created_at'] = pd.to_datetime(investments['created_at'], errors='coerce')
investments['updated_at'] = pd.to_datetime(investments['updated_at'], errors='coerce')
```

```python
ipos['public_at']    = pd.to_datetime(ipos['public_at'],    errors='coerce')
ipos['created_at']   = pd.to_datetime(ipos['created_at'],   errors='coerce')
ipos['updated_at']   = pd.to_datetime(ipos['updated_at'],   errors='coerce')
```

```python
milestones['milestone_at'] = pd.to_datetime(milestones['milestone_at'], errors='coerce')
milestones['created_at']   = pd.to_datetime(milestones['created_at'],   errors='coerce')
milestones['updated_at']   = pd.to_datetime(milestones['updated_at'],   errors='coerce')
```

```python
date_cols_objects = ['founded_at', 'closed_at', 'first_investment_at', 'last_investment_at', 
                     'first_funding_at', 'last_funding_at', 'first_milestone_at', 'last_milestone_at', 
                     'created_at', 'updated_at']

for col in date_cols_objects:
    objects[col] = pd.to_datetime(objects[col], errors='coerce')
```

```python
relationships['start_at']    = pd.to_datetime(relationships['start_at'],    errors='coerce')
relationships['end_at']      = pd.to_datetime(relationships['end_at'],      errors='coerce')
relationships['created_at']  = pd.to_datetime(relationships['created_at'],  errors='coerce')
relationships['updated_at']  = pd.to_datetime(relationships['updated_at'],  errors='coerce')
relationships['is_past']     = relationships['is_past'].astype(bool)
```

## **Final Verification**

```python
print("\n" + "=" * 80)
print("Final Summary After Cleaning")
print("=" * 80)

for name, df in dataset.items():
    total_missing = df.isnull().sum().sum()
    print(f"\n {name}")
    print(f"   Shape        : {df.shape}")
    print(f"   Total Missing: {total_missing}")
    print(f"   Duplicates   : {df.duplicated().sum()}")
```
